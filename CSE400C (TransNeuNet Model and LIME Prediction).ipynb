# Balance Class

```python
print("Original dataset shape:", df.shape)

# ---------------------------
# Step 2: Balance dataset
# ---------------------------
target_size = 5000
balanced_dfs = []

for category, group in df.groupby('category'):
    if len(group) > target_size:
        sampled = group.iloc[-target_size:]  # keep last rows
    else:
        sampled = group
    balanced_dfs.append(sampled)

df = pd.concat(balanced_dfs, ignore_index=True)
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

print("Balanced dataset shape:", df.shape)
print(df['category'].value_counts())
```

```python
from collections import Counter
from itertools import tee
import pandas as pd

# ---------------------------
# Function to generate bigrams from a list of tokens
# ---------------------------
def generate_bigrams(tokens):
    a, b = tee(tokens)
    next(b, None)
    return zip(a, b)

# ---------------------------
# Dictionary to store top bigrams per balanced category
# ---------------------------
category_bigram_freq = {}

for category in df['category'].unique():  # df should be the balanced dataset
    tokens_lists = df[df['category'] == category]['token_list']
    
    all_bigrams = []
    for token_list in tokens_lists:
        bigrams = generate_bigrams(token_list)
        all_bigrams.extend([' '.join(pair) for pair in bigrams])
    
    # Count bigrams
    bigram_counter = Counter(all_bigrams)
    
    # Get top 30 bigrams
    top_30_bigrams = bigram_counter.most_common(30)
    category_bigram_freq[category] = top_30_bigrams

# ---------------------------
# Print top 30 bigrams per category
# ---------------------------
for category, bigrams in category_bigram_freq.items():
    print(f"\n🔸 Top 2-word bigrams in category: {category}")
    for bigram, freq in bigrams:
        print(f"{bigram}: {freq}")

# ---------------------------
# Create DataFrame for plotting / analysis
# ---------------------------
bigram_rows = []
for category, bigrams in category_bigram_freq.items():
    for bigram, freq in bigrams:
        bigram_rows.append({
            'category': category,
            'bigram': bigram,
            'frequency': freq
        })

bigram_df = pd.DataFrame(bigram_rows)
```

# Temporal Insights

```python
import re
from datetime import datetime

# Bangla month mapping
BN_MONTHS = {
    "জানুয়ারি": 1,
    "ফেব্রুয়ারি": 2,
    "মার্চ": 3,
    "এপ্রিল": 4,
    "মে": 5,
    "জুন": 6,
    "জুলাই": 7,
    "আগস্ট": 8,
    "সেপ্টেম্বর": 9,
    "অক্টোবর": 10,
    "নভেম্বর": 11,
    "ডিসেম্বর": 12
}

# Bangla digit mapping
BN_DIGITS = str.maketrans("০১২৩৪৫৬৭৮৯", "0123456789")

def parse_bangla_date(date_str: str):
    """
    Parse Bangla date strings like:
      - "২৭ ডিসেম্বর ২০১৭, ১২:৩৬"
      - "প্রকাশিত: ১৪:১৫, ৩ অক্টোবর ২০২৩"
    Returns datetime object if parsed, else None.
    """
    if not isinstance(date_str, str):
        return None
    
    # remove leading words like 'প্রকাশিত:'
    cleaned = re.sub(r"^[^\d\u09E6-\u09EF]+", "", date_str.strip())
    
    # convert Bangla digits -> English
    cleaned = cleaned.translate(BN_DIGITS)

    # regex pattern (day month year, optional time OR time, day month year)
    pattern1 = re.compile(r"(?P<day>\d{1,2})\s+(?P<month>[^\s]+)\s+(?P<year>\d{4})(?:,\s*(?P<time>\d{1,2}:\d{2}))?")
    pattern2 = re.compile(r"(?P<time>\d{1,2}:\d{2}),\s*(?P<day>\d{1,2})\s+(?P<month>[^\s]+)\s+(?P<year>\d{4})")

    match = pattern1.search(cleaned) or pattern2.search(cleaned)
    if not match:
        return None

    gd = match.groupdict()

    day = int(gd["day"])
    month = BN_MONTHS.get(gd["month"], None)
    year = int(gd["year"])
    
    if month is None:
        return None
    
    # default time if missing
    if gd.get("time"):
        time_part = gd["time"]
    else:
        time_part = "00:00"
    
    try:
        return datetime.strptime(f"{day:02d}-{month:02d}-{year} {time_part}", "%d-%m-%Y %H:%M")
    except ValueError:
        return None


# ------------------ test it ------------------
examples = [
    "২৭ ডিসেম্বর ২০১৭, ১২:৩৬",
    "প্রকাশিত: ১৪:১৫, ৩ অক্টোবর ২০২৩",
    "৫ মে ২০২১"
]

for ex in examples:
    print(ex, "→", parse_bangla_date(ex))
# assuming df is your dataframe and parse_bangla_date is already defined

# create new datetime column
df["datetime"] = df["published_date"].apply(parse_bangla_date)

# optional: check how many failed parses
print("Unparsed rows:", df["datetime"].isna().sum())

#print(df[["published_date", "datetime"]].head(10))
```

```python
import torch
from transformers import AutoTokenizer, AutoModel, T5EncoderModel
from tqdm import tqdm

# -------------------------------
# Step 1: Load Bangla BERT
# -------------------------------
bert_model_name = "sagorsarker/bangla-bert-base"
bert_tokenizer = AutoTokenizer.from_pretrained(bert_model_name)
bert_model = AutoModel.from_pretrained(bert_model_name)
bert_model.eval()  # evaluation mode

# -------------------------------
# Step 2: Prepare cleaned content
# -------------------------------
balanced_texts = df['cleaned_content'].tolist()  # cleaned texts from balanced dataframe

# -------------------------------
# Step 3: Function to get BERT embeddings (CLS token)
# -------------------------------
def get_bert_embeddings(text_list, batch_size=32, device='cpu'):
    all_embeddings = []

    bert_model.to(device)
    for i in tqdm(range(0, len(text_list), batch_size)):
        batch_texts = text_list[i:i+batch_size]
        inputs = bert_tokenizer(batch_texts, padding=True, truncation=True, return_tensors="pt", max_length=256)
        inputs = {k: v.to(device) for k, v in inputs.items()}

        with torch.no_grad():
            outputs = bert_model(**inputs)
            # CLS token embedding (first token)
            batch_embeddings = outputs.last_hidden_state[:, 0, :].cpu()
            all_embeddings.append(batch_embeddings)

    all_embeddings = torch.cat(all_embeddings, dim=0)
    return all_embeddings

# -------------------------------
# Step 4: Compute BERT embeddings
# -------------------------------
bert_embeddings = get_bert_embeddings(
    balanced_texts,
    batch_size=64,
    device='cuda' if torch.cuda.is_available() else 'cpu'
)
print("BERT Embeddings shape:", bert_embeddings.shape)
```

```python
# -------------------------------
# Step 5: Load Bangla T5 (mT5)
# -------------------------------
t5_model_name = "google/mt5-base"
t5_tokenizer = AutoTokenizer.from_pretrained(t5_model_name)
t5_model = T5EncoderModel.from_pretrained(t5_model_name)
t5_model.eval()

# -------------------------------
# Step 6: Function to get T5 embeddings (mean pooling)
# -------------------------------
def get_t5_embeddings(text_list, batch_size=32, device='cpu'):
    all_embeddings = []

    t5_model.to(device)
    for i in tqdm(range(0, len(text_list), batch_size)):
        batch_texts = text_list[i:i+batch_size]
        inputs = t5_tokenizer(batch_texts, padding=True, truncation=True, return_tensors="pt", max_length=256)
        inputs = {k: v.to(device) for k, v in inputs.items()}

        with torch.no_grad():
            outputs = t5_model(**inputs)
            hidden_states = outputs.last_hidden_state  # (batch, seq_len, hidden_size)
            
            # Mean pooling (ignores padding)
            attention_mask = inputs['attention_mask'].unsqueeze(-1)  # (batch, seq_len, 1)
            sum_hidden = torch.sum(hidden_states * attention_mask, dim=1)
            lengths = torch.clamp(attention_mask.sum(dim=1), min=1e-9)
            batch_embeddings = (sum_hidden / lengths).cpu()
            all_embeddings.append(batch_embeddings)

    all_embeddings = torch.cat(all_embeddings, dim=0)
    return all_embeddings

# -------------------------------
# Step 7: Compute T5 embeddings
# -------------------------------
t5_embeddings = get_t5_embeddings(
    balanced_texts,
    batch_size=64,
    device='cuda' if torch.cuda.is_available() else 'cpu'
)
print("T5 Embeddings shape:", t5_embeddings.shape)
```

```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import LabelEncoder, label_binarize
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# -------------------------
# 1️⃣ Dataset Class for ANN (no extra dims)
# -------------------------
class NewsDataset(Dataset):
    def __init__(self, bert_embeddings, t5_embeddings, labels):
        self.bert = torch.tensor(bert_embeddings, dtype=torch.float32)
        self.t5 = torch.tensor(t5_embeddings, dtype=torch.float32)
        self.labels = torch.tensor(labels, dtype=torch.long)
        
    def __len__(self):
        return len(self.labels)
    
    def __getitem__(self, idx):
        bert_feat = self.bert[idx]  # shape: (bert_dim,)
        t5_feat = self.t5[idx]      # shape: (t5_dim,)
        label = self.labels[idx]
        return bert_feat, t5_feat, label

# -------------------------
# 2️⃣ TransformNeuNet using ANN
# -------------------------
class TransformNeuNetANN(nn.Module):
    def __init__(self, bert_dim=768, t5_dim=768, num_classes=10, hidden_dim=256):
        super().__init__()
        # BERT branch
        self.bert_fc = nn.Sequential(
            nn.Linear(bert_dim, hidden_dim),
            nn.ReLU(),
            nn.Dropout(0.3)
        )
        # T5 branch
        self.t5_fc = nn.Sequential(
            nn.Linear(t5_dim, hidden_dim),
            nn.ReLU(),
            nn.Dropout(0.3)
        )
        # Combined fully connected layers
        self.fc = nn.Sequential(
            nn.Linear(hidden_dim*2, 128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, num_classes)
        )
        
    def forward(self, bert_feat, t5_feat):
        bert_out = self.bert_fc(bert_feat)
        t5_out = self.t5_fc(t5_feat)
        combined = torch.cat([bert_out, t5_out], dim=1)
        out = self.fc(combined)
        return out

# -------------------------
# 3️⃣ Prepare Data
# -------------------------
labels = df['category'].values
le = LabelEncoder()
y = le.fit_transform(labels)

X_train_idx, X_test_idx, y_train, y_test = train_test_split(
    np.arange(len(y)), y, test_size=0.2, stratify=y, random_state=42
)

train_dataset = NewsDataset(bert_embeddings[X_train_idx],
                            t5_embeddings[X_train_idx],
                            y_train)
test_dataset = NewsDataset(bert_embeddings[X_test_idx],
                           t5_embeddings[X_test_idx],
                           y_test)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# -------------------------
# 4️⃣ Train TransNeuNet
# -------------------------
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = TransformNeuNetANN(bert_dim=bert_embeddings.shape[1],
                           t5_dim=t5_embeddings.shape[1],
                           num_classes=len(le.classes_))
model.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
epochs = 50

train_losses, val_losses, train_accs, val_accs = [], [], [], []

for epoch in range(epochs):
    # Training
    model.train()
    running_loss, correct, total = 0, 0, 0
    for bert_feat, t5_feat, labels_batch in train_loader:
        bert_feat, t5_feat, labels_batch = bert_feat.to(device), t5_feat.to(device), labels_batch.to(device)
        
        optimizer.zero_grad()
        outputs = model(bert_feat, t5_feat)
        loss = criterion(outputs, labels_batch)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item() * labels_batch.size(0)
        preds = outputs.argmax(dim=1)
        correct += (preds == labels_batch).sum().item()
        total += labels_batch.size(0)
    
    train_losses.append(running_loss / total)
    train_accs.append(correct / total)
    
    # Validation
    model.eval()
    val_loss, correct, total = 0, 0, 0
    with torch.no_grad():
        for bert_feat, t5_feat, labels_batch in test_loader:
            bert_feat, t5_feat, labels_batch = bert_feat.to(device), t5_feat.to(device), labels_batch.to(device)
            outputs = model(bert_feat, t5_feat)
            loss = criterion(outputs, labels_batch)
            
            val_loss += loss.item() * labels_batch.size(0)
            preds = outputs.argmax(dim=1)
            correct += (preds == labels_batch).sum().item()
            total += labels_batch.size(0)
    
    val_losses.append(val_loss / total)
    val_accs.append(correct / total)
    
    print(f"Epoch {epoch+1}/{epochs} → "
          f"Train Loss: {train_losses[-1]:.4f}, Val Loss: {val_losses[-1]:.4f}, "
          f"Train Acc: {train_accs[-1]:.4f}, Val Acc: {val_accs[-1]:.4f}")

# -------------------------
# 5️⃣ Evaluation Class
# -------------------------
class ModelEvaluator:
    def __init__(self, model_name, y_true, y_pred, y_score=None, le=None):
        self.model_name = model_name
        self.y_true = y_true
        self.y_pred = y_pred
        self.y_score = y_score
        self.le = le
        self.num_classes = len(np.unique(y_true))
        
    def confusion_matrix(self):
        cm = confusion_matrix(self.y_true, self.y_pred)
        plt.figure(figsize=(10,8))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=self.le.classes_ if self.le else None,
                    yticklabels=self.le.classes_ if self.le else None)
        plt.xlabel("Predicted")
        plt.ylabel("Actual")
        plt.title(f"Confusion Matrix of {self.model_name}")
        plt.xticks(rotation=45)
        plt.yticks(rotation=0)
        plt.show()
        
    def classification_report(self):
        report = classification_report(self.y_true, self.y_pred,
                                       target_names=self.le.classes_ if self.le else None)
        print(f"Classification Report for {self.model_name}:\n", report)
    
    def metrics(self):
        acc = accuracy_score(self.y_true, self.y_pred)
        prec = precision_score(self.y_true, self.y_pred, average='weighted')
        rec = recall_score(self.y_true, self.y_pred, average='weighted')
        f1 = f1_score(self.y_true, self.y_pred, average='weighted')
        print(f"{self.model_name} Metrics → Accuracy: {acc:.4f}, Precision: {prec:.4f}, Recall: {rec:.4f}, F1: {f1:.4f}")
    
    def loss_accuracy_curve(self, train_losses, val_losses, train_accs=None, val_accs=None):
        plt.figure(figsize=(10,6))
        plt.plot(range(1, len(train_losses)+1), train_losses, marker='o', label="Train Loss")
        plt.plot(range(1, len(val_losses)+1), val_losses, marker='s', label="Validation Loss")
        if train_accs is not None and val_accs is not None:
            plt.plot(range(1, len(train_accs)+1), train_accs, marker='o', linestyle='--', label="Train Acc")
            plt.plot(range(1, len(val_accs)+1), val_accs, marker='s', linestyle='--', label="Val Acc")
        plt.xlabel("Epoch")
        plt.ylabel("Loss / Accuracy")
        plt.title(f"Training & Validation Curve of {self.model_name}")
        plt.grid(True)
        plt.legend()
        plt.show()
    
    def roc_curve_multi(self):
        if self.y_score is None or self.le is None:
            print("y_score and LabelEncoder required for ROC curve")
            return
        y_true_bin = label_binarize(self.y_true, classes=np.arange(self.num_classes))
        plt.figure(figsize=(10,8))
        for i in range(self.num_classes):
            fpr, tpr, _ = roc_curve(y_true_bin[:, i], self.y_score[:, i])
            roc_auc = auc(fpr, tpr)
            plt.plot(fpr, tpr, lw=2, label=f"{self.le.classes_[i]} (AUC = {roc_auc:.2f})")
        plt.plot([0,1], [0,1], color='navy', lw=2, linestyle='--')
        plt.xlim([0.0,1.0])
        plt.ylim([0.0,1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f"ROC Curve for {self.model_name}")
        plt.legend(loc="lower right")
        plt.grid(True)
        plt.show()

# -------------------------
# 6️⃣ Predict & Evaluate
# -------------------------
model.eval()
y_preds, y_scores = [], []

with torch.no_grad():
    for bert_feat, t5_feat, labels_batch in test_loader:
        bert_feat = bert_feat.to(device)
        t5_feat = t5_feat.to(device)
        outputs = model(bert_feat, t5_feat)
        y_preds.extend(outputs.argmax(dim=1).cpu().numpy())
        y_scores.extend(torch.softmax(outputs, dim=1).cpu().numpy())

y_preds = np.array(y_preds)
y_scores = np.array(y_scores)

evaluator = ModelEvaluator("TransNeuNet", y_test, y_preds, y_score=y_scores, le=le)
evaluator.metrics()
evaluator.confusion_matrix()
evaluator.classification_report()
evaluator.loss_accuracy_curve(train_losses, val_losses, train_accs, val_accs)
evaluator.roc_curve_multi()
```

```python
import torch
import numpy as np
import pandas as pd
import os
from lime.lime_text import LimeTextExplainer
import plotly.express as px

# -------------------------
# 7️⃣ Live Prediction Function (Updated for ANN)
# -------------------------
def live_predict(text, model, bert_tokenizer, bert_model, t5_tokenizer, t5_model, le, device='cpu'):
    model.eval()
    bert_model.to(device)
    t5_model.to(device)
    
    with torch.no_grad():
        # ---- BERT embedding (CLS token) ----
        bert_inputs = bert_tokenizer([text], padding=True, truncation=True,
                                     return_tensors="pt", max_length=256).to(device)
        bert_out = bert_model(**bert_inputs).last_hidden_state[:, 0, :]  # shape: (1, bert_dim)
        
        # ---- T5 embedding (mean pooling with attention mask) ----
        t5_inputs = t5_tokenizer([text], padding=True, truncation=True,
                                 return_tensors="pt", max_length=256).to(device)
        t5_hidden = t5_model(**t5_inputs).last_hidden_state  # (1, seq_len, dim)
        attention_mask = t5_inputs['attention_mask'].unsqueeze(-1)  # (1, seq_len, 1)
        t5_out = (torch.sum(t5_hidden * attention_mask, dim=1) /
                  torch.clamp(attention_mask.sum(dim=1), min=1e-9))  # (1, t5_dim)
        
        # ---- ANN Model prediction ----
        output = model(bert_out, t5_out)
        pred_idx = output.argmax(dim=1).item()
        pred_class = le.inverse_transform([pred_idx])[0]
        pred_probs = torch.softmax(output, dim=1).cpu().numpy()[0]
        
    return pred_class, pred_probs

# =========================
# LIME Plot Function (Top 15 words, green/red) - Compact Version
# =========================
def plot_lime_explanation(exp, pred_idx, le, top_n=15):
    font_path = "/kaggle/input/bangaltextfont/NotoSansBengali-Regular.ttf"
    bangla_font = "Noto Sans Bengali" if os.path.exists(font_path) else "Nikosh"

    explanation = exp.as_list(label=pred_idx)
    df = pd.DataFrame(explanation, columns=['token', 'weight'])
    
    df = df.reindex(df['weight'].abs().sort_values(ascending=False).index)
    df = df.head(top_n)
    df['color'] = df['weight'].apply(lambda w: 'green' if w > 0 else 'red')

    fig = px.bar(
        df, x='weight', y='token', orientation='h',
        color='color', color_discrete_map={'green': 'green', 'red': 'red'},
        text='weight'
    )
    fig.update_layout(
        width=1200,   # 🔹 reduced width
        height=1000,  # 🔹 reduced height for 15 words
        yaxis={'categoryorder': 'total ascending'},
        font=dict(family=bangla_font, size=16),
        title=f"Top {top_n} Influential Words (LIME Explanation)"
    )
    fig.show()

# =========================
# LIME Wrapper Class
# =========================
class LIMEWrapper:
    def __init__(self, model, bert_tokenizer, bert_model, t5_tokenizer, t5_model, le, device='cpu'):
        self.model = model
        self.bert_tokenizer = bert_tokenizer
        self.bert_model = bert_model
        self.t5_tokenizer = t5_tokenizer
        self.t5_model = t5_model
        self.le = le
        self.device = device
        self.model.eval(); self.bert_model.eval(); self.t5_model.eval()

    def predict_proba(self, texts):
        probs = []
        with torch.no_grad():
            for text in texts:
                bert_in = self.bert_tokenizer([text], padding=True, truncation=True,
                                              return_tensors="pt", max_length=256).to(self.device)
                bert_out = self.bert_model(**bert_in).last_hidden_state[:, 0, :]
                
                t5_in = self.t5_tokenizer([text], padding=True, truncation=True,
                                          return_tensors="pt", max_length=256).to(self.device)
                t5_hidden = self.t5_model(**t5_in).last_hidden_state
                att_mask = t5_in['attention_mask'].unsqueeze(-1)
                t5_out = (torch.sum(t5_hidden * att_mask, dim=1) /
                          torch.clamp(att_mask.sum(dim=1), min=1e-9))

                out = self.model(bert_out, t5_out)
                prob = torch.softmax(out, dim=1).cpu().numpy()[0]
                probs.append(prob)
        return np.array(probs)

# =========================
# LIME Live Predict (Bangla word-level)
# =========================
def lime_live_predict(text, model, bert_tokenizer, bert_model, t5_tokenizer, t5_model, le, device='cpu'):
    wrapper = LIMEWrapper(model, bert_tokenizer, bert_model, t5_tokenizer, t5_model, le, device)

    pred_class, pred_probs = live_predict(text, model, bert_tokenizer, bert_model, t5_tokenizer, t5_model, le, device)
    pred_idx = np.argmax(pred_probs)

    # ---- Bangla word-level tokenizer for LIME ----
    def bangla_tokenizer(text):
        return text.split()  # simple whitespace split

    explainer = LimeTextExplainer(class_names=le.classes_, split_expression=bangla_tokenizer)
    exp = explainer.explain_instance(text, wrapper.predict_proba, num_features=15, top_labels=1)

    # Plot top 15 words
    plot_lime_explanation(exp, pred_idx, le, top_n=15)

    # Print top 15 influential words
    important_words = exp.as_list(label=pred_idx)
    important_words = sorted(important_words, key=lambda x: abs(x[1]), reverse=True)[:15]

    print("\n📝 Most Influential Words for Prediction:")
    for token, weight in important_words:
        effect = "↑ Positive" if weight > 0 else "↓ Negative"
        print(f"{token:15s} | {weight:.4f} | {effect}")

    return pred_class, pred_probs, exp

# =========================
# Example Usage
# =========================
user_text = "জাতীয় নাগরিক পার্টির (এনসিপি) উত্তরাঞ্চলের মুখ্য সংগঠক সারজিস আলম বলেছেন, ‘আমরা স্পষ্ট করে বলব, বাংলাদেশের মানুষ ওই একটা প্রশ্নে আপসহীন, শেখ হাসিনা খুনি—বাংলাদেশের দেয়ালে দেয়ালে এই লেখাটা রক্ত দিয়ে লেখা হয়ে গেছে। যাঁরা এখন অন্তর্বর্তী সরকারের সুবিধাভোগী, তাঁরা তাঁদের জায়গা থেকে এইটা ভুলে যেতে পারেন, কিন্তু যে মা তাঁর ছেলের লাশ নিজের হাতে ছুঁয়েছেন, ওই মা কিন্তু এটা ভুলতে পারবেন না। ওই মা যত দিন না ভুলবেন, আমরা যাঁরা এই অভ্যুত্থানে রাজপথে লড়াই করেছিলাম, আমরা এটা ভুলব না, ভুলতে দেবও না।’ পঞ্চগড়ের আটোয়ারী উপজেলার বিভিন্ন পূজামণ্ডপ পরিদর্শনের সময় বুধবার বিকেলে মালিগাঁও তেঁতুলতলা দুর্গামন্দিরে সাংবাদিকদের এ কথা বলেন সারজিস। তিনি বলেন, ‘আমরা স্পষ্ট করে বলি এই বাংলাদেশে আওয়ামী লীগের আর কোনো রাজনীতি চলবে না।’ এনসিপির এই নেতা বলেন, ‘অভ্যুত্থান–পরবর্তী সময়ে যখন আমাদের আবার আওয়ামী লীগ নিষিদ্ধের দাবিতে রাজপথে নামতে হয়, এটাই হচ্ছে এই অন্তর্বর্তীকালীন সরকারের প্রথম ব্যর্থতা। দ্বিতীয়ত, যখন মানুষ আবার রাজপথে নামল তাঁদের জায়গা থেকে এবং নামার পরে মানুষ রাজপথে দিনের পর দিন আন্দোলন করল এবং যমুনা ঘেরাওয়ের পর্যায় পর্যন্ত গেল, তখন তারা এই রাজনৈতিক কার্যক্রম নিষিদ্ধের যে ধাপ সেই ধাপে গেল।’ প্রধান উপদেষ্টা অধ্যাপক মুহাম্মদ ইউনূসের উদ্দেশে সারজিস আলম বলেন, ‘তিনি যদি এখন বিভিন্ন কারণে আন্তর্জাতিক মহলে বাংলাদেশের মানুষের এত ত্যাগ–তিতিক্ষা সবকিছু ভুলে গিয়ে, বাংলাদেশের মানুষের এত রক্ত বিসর্জন-প্রাণ বিসর্জন ভুলে গিয়ে নিজের ইমেজের চেক অ্যান্ড ব্যালেন্স করার জন্য সুশীল সাজতে শুরু করেন এবং এই অভ্যুত্থানের মতো একটা বড় ঘটনা ভুলে গিয়ে তিনি যদি এই স্টেটমেন্ট আসলে দিয়ে থাকেন যে আওয়ামী লীগের কার্যক্রম আবার সচল করা হতে পারে, তাহলে আমরা শুধু একটা কথাই বলি, বাংলাদেশের মানুষ ছাড় দিতে পারে, কিন্তু ছেড়ে দেবে না।’"

pred_class, pred_probs, lime_exp = lime_live_predict(
    user_text, model, bert_tokenizer, bert_model, t5_tokenizer, t5_model, le,
    device='cuda' if torch.cuda.is_available() else 'cpu'
)

print(f"\nPredicted Class: {pred_class}")
print("Class Probabilities:")
for cls, prob in zip(le.classes_, pred_probs):
    print(f"{cls}: {prob:.4f}")
```
